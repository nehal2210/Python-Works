## scrapping of side bar of html side bar of W3School

```python
url="https://www.w3schools.com/"
import requests
from bs4 import BeautifulSoup
res=requests.get(url)
res

# soup=Beautifulsoup()
```

```python
# google cloud platform

soup=BeautifulSoup(res.text,"html.parser")
print(soup)
```

```python
#ache formate me agr likhna hhe to.
print(soup.prettify())
```

```python
nav=soup.nav
print(nav)
```

```python
# convert it into string to write ile
nav_string=str(nav)
print(nav_string)
```

```python
# ab me es ki css dhoondne lga hun.
print(soup.text) #mujhe pata chla k css nhi nikaal skte
```

```python

url_css=requests.get("https://www.w3schools.com/w3css/4/w3.css")
soupcss=BeautifulSoup(url_css.text,"html.parser")
```

```python
print(soupcss.text)
```

## whatsapp your contact extract application

```python
from bs4 import BeautifulSoup
from requests import get
payload = {'t': 's', 'u':'923352155334-1536594720@g.us','i':'1536866084'}
url="https://web.whatsapp.com"
response=get(url,params=payload) # request send
response
```

```python
soupw=BeautifulSoup(response.text,"html.parser") #object of beautiful soup
```

```python
print(soupw)
```

```python
print("asf")
```

```python
print(soupw.prettify())
```
